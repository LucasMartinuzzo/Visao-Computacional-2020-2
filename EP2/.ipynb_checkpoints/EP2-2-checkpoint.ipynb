### Funções Auxiliares

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from matplotlib.colors import NoNorm
from scipy.ndimage import convolve
from skimage import exposure
from PIL import Image
import os
import shutil
import numpy as np
import cv2 as cv
%matplotlib inline
```

```python
original_directory = "../Imagens/OriginalDataset/"
original_gray_dir = '../Imagens/OriginalGrayDataset/'
augmented_dir = '../Imagens/AugmentedDataset/'
normalized_dir = '../Imagens/NormalizedDataset/'
averaged_dir = '../Imagens/AveragedDataset/'
IMAGE_SHAPE = (960,720)
N_PIXELS = IMAGE_SHAPE[0]*IMAGE_SHAPE[1]
N_CLASSES = 10
N_SAMPLE_BY_CLASS = 4
```

```python
def selectRandImages(image_dir,n_sample_by_class = 4):
    folders = os.listdir(image_dir)
    rand_images = {}
    for folder in folders:
        list_images = []
        files = os.listdir(os.path.join(image_dir,folder))
        for sample in range(n_sample_by_class):
            list_images.append(files[np.random.randint(len(files),size=1)[0]])
            list_images.append(files[np.random.randint(len(files),size=1)[0]])
            list_images.append(files[np.random.randint(len(files),size=1)[0]])
            list_images.append(files[np.random.randint(len(files),size=1)[0]])
        rand_images[folder] = list_images
    return rand_images
```

```python
def plotSample(image_dir,sample_images_dict,columns=4,gray_mode=False,prefix=None):
    rows = len(rand_sample)
    fig, axes = plt.subplots(rows,columns,figsize=(18,24),constrained_layout=True)
    axes.flatten()
    for i,title in enumerate(sample_images_dict):
        class_dir = os.path.join(image_dir,title)
        for j in range(columns):
            if prefix is None:
                filename = sample_images_dict[title][j]
            else:
                filename = prefix + '_' + sample_images_dict[title][j]
            image=cv.imread(os.path.join(class_dir,filename),0)
            if(gray_mode):
                axes[i][j].imshow(image,cmap='gray',norm=NoNorm())
            else:
                axes[i][j].imshow(image)
            axes[i,0].set_ylabel(title,rotation=90,size='large',color="red")
            axes[i][j].set_yticks([])
            axes[i][j].set_xticks([])
    fig.show()
    return
```

```python
def applyFilter(source_dir,target_dir,function,prefix=None,aux=None):
    if not os.path.exists(target_dir):
        os.makedirs(target_dir)
    classes_dir = os.listdir(source_dir)
    i = 0
    for class_folder in classes_dir:
        target_class_dir = os.path.join(target_dir,class_folder)
        if not os.path.exists(target_class_dir):
            os.makedirs(target_class_dir)
        source_class_dir = os.path.join(source_dir,class_folder)
        files = os.listdir(source_class_dir)
        for f in files:
            image = mpimg.imread(os.path.join(source_class_dir,f))
            if aux is None:
                filtered_image = Image.fromarray(function(image))
            else:
                filtered_image = Image.fromarray(function(image,aux))
            if prefix is None:
                target_filename = f
            else:
                target_filename = prefix + '_' + f
            full_filename = os.path.join(target_class_dir,target_filename)
            #print(full_filename)
            filtered_image.save(full_filename, 'JPEG', quality=95)
            i+=1
    print("Finalizado. {} imagens foram processadas.".format(i))
    return
```

```python
def plotPrototipeAndHistograms(images_dict,n_rows=N_CLASSES,plot_image_mode=True):
    n_columns = len(images_dict)
    fig, axes = plt.subplots(n_rows,n_columns,figsize=(18,24),constrained_layout=True)
    axes.flatten()
    for j,title in enumerate(images_dict):
        sub_dict = images_dict[title]
        for i, class_name in enumerate(sub_dict):
            var = sub_dict[class_name]
            if plot_image_mode:
                axes[i][j].imshow(var,cmap='gray',norm=NoNorm())
                axes[i][j].set_yticks([])
                axes[i][j].set_xticks([])
            else:
                axes[i][j].plot(var)
                #axes[i][j].set_yscale('log')
            axes[i,0].set_ylabel(class_name,rotation=90,size='large',color="red")
            
        axes[0][j].set_title(title,size='large',color="red")
    fig.show()
    return
```

#### Normalização

```python
def equalizeHist(hist,num_pixels,gray_levels=256):
    intensity_level = np.full(hist.shape,0)
    equalized_hist = np.full(hist.shape,0.0)
    for i in range(gray_levels):
        intensity_level[i][0] = (gray_levels-1)*hist[:(i+1)].sum()/num_pixels
    for i in range(gray_levels):
        equalized_hist[intensity_level[i][0]][0] += hist[i][0]
    return equalized_hist,intensity_level
```

```python
def equalizeImageUsingIntensityLevel(image,intensity_level_hist,gray_levels=256):
    equalized_image = np.full(image.shape,0)
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            equalized_image[i][j] = intensity_level_hist[image[i][j]][0]
    return equalized_image.astype(np.uint8)
```

```python
def fullEqualizeProcess(image,gray_levels=256,return_hist=False):
    hist = cv.calcHist([image],[0],None,[gray_levels],[0,gray_levels])
    num_pixels=image.shape[0]*image.shape[1]
    eq_hist, il_hist = equalizeHist(hist,num_pixels,gray_levels)
    eq_image = equalizeImageUsingIntensityLevel(image,il_hist,gray_levels).astype(np.uint8)
    if return_hist:
        return eq_image,hist,il_hist,eq_hist
    return eq_image
```

### Protótipo e Histogramas

Nesta sessão serão construídos:

2. Histograma médio de cada classe
3. Variância do histograma de cada classe

#### Protótipo Médio de cada classe

```python
def createAveragedImage(source_dir,target_dir,shape,prefix):
    if not os.path.exists(target_dir):
        os.makedirs(target_dir)
    classes_dir = os.listdir(source_dir)
    i = 0
    averaged_image_dict = {}
    for class_folder in classes_dir:
        target_class_dir = os.path.join(target_dir,class_folder)
        if not os.path.exists(target_class_dir):
            os.makedirs(target_class_dir)
        source_class_dir = os.path.join(source_dir,class_folder)
        files = os.listdir(source_class_dir)
        averaged_image = np.full(shape,0.0)
        n_images = 0
        for f in files:
            image = mpimg.imread(os.path.join(source_class_dir,f))
            averaged_image += image
            n_images += 1
        averaged_image = (averaged_image/n_images).astype(np.uint8)
        averaged_image = Image.fromarray(averaged_image)
        averaged_image_dict[class_folder] = averaged_image
        full_filename = os.path.join(target_class_dir,prefix + '_'+ 'averagedImage.jpeg')
        averaged_image.save(full_filename, 'JPEG', quality=95)
        i+=1
    print("Finalizado. {} protótipos foram criados.".format(i))
    return averaged_image_dict
```

```python
averaged_images = {}
averaged_images['normalized'] = createAveragedImage(normalized_dir,averaged_dir,IMAGE_SHAPE,'normalized')
averaged_images['originalGray'] = createAveragedImage(original_gray_dir,averaged_dir,IMAGE_SHAPE,'originalGray')
averaged_images['augmented'] = createAveragedImage(augmented_dir,averaged_dir,IMAGE_SHAPE,'augmented')
```

```python
plotPrototipeAndHistograms(averaged_images,plot_image_mode=True)
```

#### Histograma médio de cada classe
Média dos histogramas de cada classe

```python
def getAllHistograms(source_dir,num_pixels=N_PIXELS):
    classes_dir = os.listdir(source_dir)
    hist_image_dict = {}
    n_images = 0
    for class_folder in classes_dir:
        source_class_dir = os.path.join(source_dir,class_folder)
        files = os.listdir(source_class_dir)
        histogram_list = []
        for f in files:
            image = mpimg.imread(os.path.join(source_class_dir,f))
            hist = cv.calcHist([image],[0],None,[256],[0,256])/num_pixels
            histogram_list.append(hist)
            n_images += 1
        hist_image_dict[class_folder] = histogram_list
    print("Finalizado. {} histogramas foram criados.".format(n_images))
    return hist_image_dict
```

```python
hist_images = {}
hist_images['augmented'] = getAllHistograms(augmented_dir)
hist_images['normalized'] = getAllHistograms(normalized_dir)
hist_images['originalGray'] = getAllHistograms(original_gray_dir)
```

```python
def getAverageHist(histogram_dict):
    average_hist_dict = {}
    n_hist = 0
    for key,item in histogram_dict.items():
        average_hist_dict[key] = np.mean(histogram_dict[key],axis=0)
        n_hist += 1
    print("Finalizado. {} histogramas médios foram calculados.".format(n_hist))
    return average_hist_dict
```

```python
mean_hist_images = {}
mean_hist_images['augmented'] = getAverageHist(hist_images['augmented'])
mean_hist_images['normalized'] = getAverageHist(hist_images['normalized'])
mean_hist_images['originalGray'] = getAverageHist(hist_images['originalGray'])
```

```python
print('O eixo x representa o nível de cinza e o eixo y a quantidade de pixels no nível de cinza dividido pelo número total de pixels da imagem.')
plotPrototipeAndHistograms(mean_hist_images,plot_image_mode=False)
```

Acima vemos que os gráficos das médias das imagens normalizadas tem um pico no valor máximo. Examinando as imagens suspeito que a causa do pico nas imagens aumentadas é a soma do gradiente aumentou muito a quantidade de branco nas images. Os picos eram ainda maiores e os histogramas das imagens aumentadas se assemelhavam muito aos histogramas das imagens normalizadas, porém eu optei por deixar o gradiente com tons mais escuros e reprocessar as imagens, desta maneira, o pico de branco não foi tão grande nas imagens aumentadas

Colocar o gradiente com tons mais escuros ajudou a distribuir um pouco melhor a quantidade de pixels com níveis de cinza diferente de 255 nas imagens normalizadas, porém ainda há um grande pico no histograma, pois o processo de normalização saturou muito as fotos tiradas durante o dia com o fundo claro que passaram pelo processo de soma de gradiente.

#### Variância dos Histogramas de cada classe

```python
def getVarianceHist(histogram_dict):
    variance_hist_dict = {}
    n_hist = 0
    for key,item in histogram_dict.items():
        variance_hist_dict[key] = np.var(histogram_dict[key],axis=0)
        n_hist += 1
    print("Finalizado. {} histogramas médios foram calculados.".format(n_hist))
    return variance_hist_dict
```

```python
var_hist_images = {}
var_hist_images['augmented'] = getVarianceHist(hist_images['augmented'])
var_hist_images['normalized'] = getVarianceHist(hist_images['normalized'])
var_hist_images['originalGray'] = getVarianceHist(hist_images['originalGray'])
```

```python
print('O eixo x representa o nível de cinza e o eixo y a quantidade de pixels no nível de cinza dividido pelo número total de pixels da imagem.')
plotPrototipeAndHistograms(var_hist_images,plot_image_mode=False)
```
