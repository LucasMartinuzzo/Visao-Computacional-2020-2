#### Amostras da Base

```python
import matplotlib.pyplot as plt
from PIL import Image
import os
import numpy as np
%matplotlib inline
```

```python
images_directory = "../Imagens_Originais/"
folders = os.listdir(images_directory)
```

```python
rand_images = {}
for folder in folders:
    files = os.listdir(images_directory+folder)
    rand_images[folder] = images_directory+folder + "/" + files[np.random.randint(len(files),size=1)[0]]
```

```python
r = 5
c = 2
fig,ax = plt.subplots(r,c,figsize=(18,24),constrained_layout=True)
ax.flatten()
for i,title in enumerate(rand_images):
    with open(rand_images[title],'rb') as f:
        image=Image.open(f)
        ax[i//c][i%c].imshow(image)
        ax[i//c][i%c].set_title(title)
        ax[i//c][i%c].axis('off')
fig.show()
```
